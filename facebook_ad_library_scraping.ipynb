# Facebook Ad Library Scraping

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
import time
import csv
```

## Enter search criteria

```python
def enter_search_criteria(search_term):
    comboboxes = driver.find_elements_by_css_selector('div[role="combobox"]')

    comboboxes[1].click()

    options = driver.find_elements_by_css_selector('div[role="heading"]')

    ad_cate_section = [x for x in options if x.text in ['Select ad category', 'é¸æ“‡å»£å‘Šé¡åˆ¥']][0]

    ad_cate_section = ad_cate_section.find_element_by_xpath('./..')

    ad_cate_section.find_elements_by_css_selector('input')[0].click()

    text_inputs = driver.find_elements_by_css_selector('input[type="text"]')

    text_input = [x for x in text_inputs if x.get_attribute('placeholder') in ['Search by keyword or advertiser', 'ä¾é—œéµå­—æˆ–å»£å‘Šå•†æœå°‹']][0]

    text_input.send_keys(search_term)

    text_input.send_keys(Keys.ENTER)
```

## Check results count

```python
def check_results_count():
    results_count_text = driver.find_element_by_css_selector('div._94cs div._7gn2 div').text

    result_count = int(''.join(re.findall('\d+', results_count_text)))
    
    return result_count
```

## Scroll down until desired count of ads appear

```python
def scroll_down_for_desired_ads_count():
    ad_info_list = []
    while len(ad_info_list) < 100:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        ad_info_list = driver.find_elements_by_css_selector('div._99s5')
    
```

## Grab the texts

```python
def save_texts():
    ad_info_list = driver.find_elements_by_css_selector('div._99s5 div.iajz466s')

    ad_info_list = [x for i, x in enumerate(ad_info_list) if i % 2 != 0]
    print('Requesting ads count:', len(ad_info_list))
    
    texts = []
    with open('ad_texts.csv', 'w') as f:
        writer = csv.writer(f)
        for ad_info in ad_info_list:
            text = ad_info.find_element_by_tag_name('div').text
            writer.writerow([text])
            
            texts.append(text)
    
    return texts
```

```python
if __name__ == '__main__':
    url = 'https://www.facebook.com/ads/library/?active_status=all&ad_type=all&country=TW&sort_data[direction]=desc&sort_data[mode]=relevancy_monthly_grouped&media_type=all'
    driver = webdriver.Chrome('Web_Scraping/chromedriver')
    driver.get(url)
    time.sleep(3)
    
    search_term = 'è¥ªå­'
    enter_search_criteria(search_term)
    time.sleep(3)
    
    results_count = check_results_count()
    print('results_count:', results_count)
    time.sleep(3)
    
    scroll_down_for_desired_ads_count()
    time.sleep(3)
    
    texts = save_texts()
    
    driver.quit()
    
    print('==========Finished==========')
    
```
